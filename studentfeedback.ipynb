```python

plt.figure(figsize=(6,4))
sns.histplot(df["Well_Versed"], bins=10, kde=True)
plt.title("Distribution: Subject Knowledge")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df["Average Score"] = df.drop(columns=["Student_ID"]).mean(axis=1)
```

```python
df.head()
```

```python
(df["Coyrse Recommendation"] >= 8).mean() * 100
```

```python
df.head()
```

```python
avg_scores = df.drop(columns=["Student_ID"]).mean().round(2)
print("Average rating for each aspect:\n", avg_scores)
```

```python
highest_aspect = avg_scores.idxmax()
print("Highest rated aspect:", highest_aspect, "with score:", avg_scores[highest_aspect])
```

```python
lowest_aspect = avg_scores.idxmin()
print("Lowest rated aspect:", lowest_aspect, "with score:", avg_scores[lowest_aspect])
```

```python
plt.figure(figsize=(12, 8))
# Plot histograms for each feedback column (ignore Student_ID)
for i, col in enumerate(df.drop(columns=["Student_ID"]).columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], bins=10, kde=False, color="skyblue")
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
recommend_count = (df["Coyrse Recommendation"] >= 8).sum()
recommend_percent = (recommend_count / len(df)) * 100
print(f"{recommend_count} students ({recommend_percent:.2f}%) recommend the course (rating  8).")
```

```python
low_ratings = (df["Explains_Concepts"] <= 4).sum()
low_percent = (low_ratings / len(df)) * 100
print(f"{low_ratings} students ({low_percent:.2f}%) gave low ratings (4) for 'Explains Concepts'.")
```

```python
df["Overall_Score"] = df.drop(columns=["Student_ID"]).mean(axis=1)
overall_avg = df["Overall_Score"].mean().round(2)
print("Overall average student satisfaction score:", overall_avg)
```

```python
correlations = df.drop(columns=["Student_ID"]).corr()["Coyrse Recommendation"].sort_values(ascending=False)
print("Correlation with Course Recommendation:\n", correlations)
```
